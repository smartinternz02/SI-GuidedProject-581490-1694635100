# Outlier Handeling

```python
sns.boxplot(df.Age)
```

```python
q1 = df.Age.quantile(0.25)
q3 = df.Age.quantile(0.75)
print(q1)
print(q3)
```

```python
IQR = q3-q1
```

```python
upper_limit = q3+1.5*IQR
upper_limit
```

```python
lower_limit = q1-1.5*IQR
lower_limit
```

```python
df.median()
```

```python
df['Age']=np.where(df['Age']>upper_limit,29.699,df['Age'])
```

```python
sns.boxplot(df.Age)
```

```python
#For Fare
sns.boxplot(df.Fare)
```

```python
qf1 = df.Fare.quantile(0.25)
qf3 = df.Fare.quantile(0.75)
print(qf1)
print(qf3)
```

```python
IQRf = qf3-qf1
```

```python
upper_limit = qf3+1.5*IQRf
upper_limit
```

```python
lower_limit = qf1-1.5*IQRf
lower_limit
```

```python
df.median()
```

```python
df['Fare']=np.where(df['Fare']>upper_limit,14.45,df['Fare'])
```

```python
sns.boxplot(df.Fare)
```
