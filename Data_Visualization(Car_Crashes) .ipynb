# Data Discription in Brief:
The above dataset we have the information about the losses and premiums of the insurance companies in different  sates of US and other and the precentages of different categorical cases of fatal accidents or fatility rate like  total accidents, during alochol consumption, no previous history of accidents,not distracted. 
now using Data visualization we can able to understand and anlyse the data.

```python

"""total accidents in all states ,by this we can know wich states
have higest number of accidents"""
plt.figure(figsize=(30, 18))
sns.barplot( x = "abbrev",y = "total",data = df,palette="viridis",)
plt.show()
```

```python
#graph between ins_losses and ins_premium
sns.scatterplot(x="ins_losses",y="ins_premium",data=df)
sns.lineplot(x="ins_losses",y="ins_premium",data=df,ci=None)
plt.show()
"""Inference : if the inc_losses of an insurance companies are higher then 
inc_premium of the policy holders wich pay insurence are higher as observe the graph """
```

```python
# Create line and scatter plots for each variable
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.lineplot(x="speeding", y="total", data=df,color='red')
sns.scatterplot(x="speeding", y="total", data=df,color='red')
plt.title("Speeding vs. Total")


plt.subplot(2, 2, 2)
sns.lineplot(x="alcohol", y="total", data=df,color='blue')
sns.scatterplot(x="alcohol", y="total", data=df,color='blue')
plt.title("Alcohol vs. Total")


plt.subplot(2, 2, 3)
sns.lineplot(x="not_distracted", y="total", data=df,color='green')
sns.scatterplot(x="not_distracted", y="total", data=df,color='green')
plt.title("Not Distracted vs. Total")


plt.subplot(2, 2, 4)
sns.lineplot(x="no_previous", y="total", data=df,color='orange')
sns.scatterplot(x="no_previous", y="total", data=df,color='orange')
plt.title("No Previous Accidents vs. Total")


plt.tight_layout()


plt.show()


"""
Inference: if the total increases speeding,alcohol,not_distracted,no_previous
will also increase these are +ve ly corelated with total accidents 
"""
```

```python
#comparing ins_losses with other varibles
sns.scatterplot(x="speeding", y="ins_losses", data=df, label="Speeding", color='red')

sns.scatterplot(x="alcohol", y="ins_losses", data=df, label="Alcohol", color='blue')

sns.scatterplot(x="not_distracted", y="ins_losses", data=df, label="Not Distracted", color='green')

sns.scatterplot(x="no_previous", y="ins_losses", data=df, label="No Previous Accidents", color='orange')
"""
Inference: inc_loss have neutral with all other varibles except ins_premium
"""
```

# Outlier Detection

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
sns.boxplot(x="speeding", data=df, color='red')
plt.title("Speeding")
plt.subplot(2, 3, 2)
sns.boxplot(x="alcohol", data=df, color='blue')
plt.title("Alcohol")
plt.subplot(2, 3, 3)
sns.boxplot(x="not_distracted", data=df, color='green')
plt.title("Not Distracted")
plt.subplot(2, 3, 4)
sns.boxplot(x="no_previous", data=df, color='orange')
plt.title("No Previous Accidents")
plt.subplot(2, 3, 5)
sns.boxplot(x="ins_losses", data=df, color='purple')
plt.title("Inc_Losses")
plt.subplot(2, 3, 6)
sns.boxplot(x="ins_premium", data=df, color='pink')
plt.title("Inc_Premium")
plt.figure(figsize=(4,3))
sns.boxplot(x="total", data=df, color='magenta')
plt.title("Total")

plt.tight_layout()

plt.show()
"""We can observe Alcohol and not_distracted attributes have outliers"""
```

```python
#boxplots ins_losses with other varibles
plt.figure(figsize=(20, 16))
plt.subplot(2, 2, 1)
sns.boxplot(x="speeding", y="ins_losses", data=df)
plt.title("Speeding vs. Insurance Losses")
plt.subplot(2, 2, 2)
sns.boxplot(x="total", y="ins_losses", data=df)
plt.title("Total vs. Insurance Losses")
plt.subplot(2, 2, 3)
sns.boxplot(x="alcohol", y="ins_losses", data=df)
plt.title("Alcohol vs. Insurance Losses")
plt.subplot(2, 2, 4)
sns.boxplot(x="not_distracted", y="ins_losses", data=df)
plt.title("Not Distracted vs. Insurance Losses")


plt.tight_layout()
plt.show()
"""
Inference: Every boxplot is perfectly symmetrical, It means
ins_losses is constant or not much related to the other attributes 
therefore heat maps also prove this point 
"""
# Outliers are detected too
```
